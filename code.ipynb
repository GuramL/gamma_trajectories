```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('gamma_trajectories_long.csv')

# Group by 'run'
groups = df.groupby('run')

plt.figure(figsize=(8, 6))

# Plot each run's gamma0 vs gamma1 sequence
for run, group in groups:
    plt.scatter(group['gamma0'], group['gamma1'], marker='o')

plt.xlabel('gamma0')
plt.ylabel('gamma1')
plt.title('Gamma0 vs Gamma1 Trajectories by Run')
plt.legend()
plt.grid(True)
plt.show()
```
